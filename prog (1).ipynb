```python
import numpy as np
import pandas as pd
import scipy as sp
import seaborn as sns; sns.set(style='whitegrid')
```

```python
data = np.genfromtxt("data.csv",delimiter=",",skip_header=1) 
n=len(data)
N=30463
data
```

```python
data_mean=data.mean() #Среднее значение
data_mean
```

```python
data_median=np.median(data) #Медиана
data_median
```

```python
a=(data - data.mean())**2
b=(a).sum()
s2=b/(n-1) #Дисперсия
s2 
```

```python
s = np.sqrt(s2) #Стандартное отклонение
s
```

```python
Q1=np.quantile(data,0.25) #1й квартиль
Q1
```

```python
Q3=np.quantile(data,0.75) #3й квартиль
Q3
```

```python
f=n/N
mean_std = s * np.sqrt(1 - f) / np.sqrt(n) #стандартная ошибка среднего
alpha = 0.836
t = sp.stats.norm.ppf((1 + alpha) / 2) #квантиль
ci_left =  data_mean - t * mean_std  #левая граница
ci_right = data_mean + t * mean_std  #правая граница

ci=[ci_left, ci_right]
ci
```

```python
outpute_table=pd.DataFrame(columns=["mean","median","var","std_value","quantile1","quantile3","CI_left","CI_right"])
outpute_table.loc[0]=[data_mean, data_median, s2, s, Q1, Q3, ci_left, ci_right]
```

```python
outpute_table.to_csv('answers.csv',index=False)
```
